```python
import numpy as np
import pandas as pd
```

```python
#import .csv dataset

data=pd.read_csv("laptop_price.csv",encoding='latin-1')
data.sample(5)
```

```python
data.shape
```

```python
data.describe
```

```python
data.info()
```

```python
#check null values 

data.isnull().sum()
```

```python
#Remove unsed id column

data=data.drop('laptop_ID',axis=1)
```

```python
data.sample(5)
```

```python
#Remove GB and kg from attributes then convert their values into numerical

data['Ram'] = data['Ram'].astype(str).str.replace('GB', '').astype('int32')
data['Weight'] = data['Weight'].astype(str).str.replace('kg', '').astype('float32')
data.sample(5)
```

```python
#Consider the Company column

data['Company'].value_counts()
from sklearn.preprocessing import LabelEncoder
obj=LabelEncoder()
result_encoded_company=obj.fit_transform(data['Company'])
result_encoded_company

from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
company_encoded_with_binarizer=obj_binarizer.fit_transform(data['Company'])
data[obj_binarizer.classes_]=company_encoded_with_binarizer
data.head(5),obj_binarizer.classes_
```

```python
data=data.drop('Company',axis=1)
data.head(5)
```

```python
#Consider the Product feature

data['Product'].nunique()
data=data.drop('Product',axis=1)
data.head(5)
```

```python
#for TypeName feature

data['TypeName'].value_counts()
from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
typename_encoded_with_binarizer=obj_binarizer.fit_transform(data['TypeName'])
data[obj_binarizer.classes_]=typename_encoded_with_binarizer
data=data.drop('TypeName',axis=1)
data.head(5),obj_binarizer.classes_
```

```python
data.head(5)
```

```python
#for Inches feature

data['Inches'].value_counts()
```

```python
#for Resolution feature

#extract the resolution's features one by one 
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x: 1 if 'Touchscreen' in x else 0)
data['IPS'] = data['ScreenResolution'].apply(lambda x: 1 if 'IPS' in x else 0)
```

```python
data.head(5)
```

```python
# Extract horizontal and vertical resolution separately
resolution_split = data['ScreenResolution'].str.extract(r'(\d+)x(\d+)')
full_resolution = resolution_split[0].astype(str) + '*' + resolution_split[1].astype(str)
from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
resolution_encoded_with_binarizer=obj_binarizer.fit_transform(full_resolution)
data[obj_binarizer.classes_]=resolution_encoded_with_binarizer
data=data.drop('ScreenResolution',axis=1)
data.head(5),obj_binarizer.classes_
```

```python
data.head(5)
```

```python
#For CPU feature

data['Cpu_Manufacturer'] = data['Cpu'].apply(lambda x: x.split()[0])  # e.g., Intel, AMD
data['Cpu_Series'] = data['Cpu'].apply(lambda x: ' '.join(x.split()[0:3])) 
data=data.drop('Cpu',axis=1)
data['Cpu_Manufacturer'].value_counts(),data['Cpu_Series'].value_counts()
```

```python
from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
cpu_Manufacturer_encoded_with_binarizer=obj_binarizer.fit_transform(data['Cpu_Manufacturer'])
for idx, class_name in enumerate(obj_binarizer.classes_): 
    data[f'CPU_{class_name}'] = cpu_Manufacturer_encoded_with_binarizer[:, idx]
data=data.drop('Cpu_Manufacturer',axis=1)
```

```python
data.head(5)
```

```python
def set_processor (name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split() [0] == 'AMD':
            return 'AMD_series_processor'
        else:
            return 'Other_series_processor'
data['Cpu_Series']= data['Cpu_Series'].apply(set_processor)
data['Cpu_Series'].value_counts()
```

```python
from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
cpu_series_encoded_with_binarizer=obj_binarizer.fit_transform(data['Cpu_Series'])
data[obj_binarizer.classes_]=cpu_series_encoded_with_binarizer
data=data.drop('Cpu_Series',axis=1)
```

```python
data.head(5)
```

```python
#for Gpu feature

data['Gpu_Manufacturer'] = data['Gpu'].apply(lambda x: x.split()[0])
data['Gpu_Manufacturer'].value_counts()                                           
```

```python
from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
gpu_series_encoded_with_binarizer=obj_binarizer.fit_transform(data['Gpu_Manufacturer'])
for idx, class_name in enumerate(obj_binarizer.classes_): 
    data[f'GPU_{class_name}'] = gpu_series_encoded_with_binarizer[:, idx]
data=data.drop('Gpu_Manufacturer',axis=1)
data=data.drop('Gpu',axis=1)
data.head(5)
```

```python
#for OpSys feature

def simplify_os(os):
    if 'Windows' in os:
        return 'Windows'
    elif 'Mac' in os:
        return 'MacOS'
    elif 'Linux' in os or 'Ubuntu' in os:
        return 'Linux'
    else:
        return 'Other/No OS'

data['OpSys'] = data['OpSys'].apply(simplify_os)
```

```python
data['OpSys'].value_counts()
```

```python
from sklearn.preprocessing import LabelBinarizer
obj_binarizer=LabelBinarizer()
os_series_encoded_with_binarizer=obj_binarizer.fit_transform(data['OpSys'])
data[obj_binarizer.classes_]=os_series_encoded_with_binarizer
data=data.drop('OpSys',axis=1)
data.head(5)
```

```python
data['Weight'].value_counts()
```

```python
x=data.drop('Price_euros',axis=1)
y=data['Price_euros']
x.shape
```

```python
x = pd.DataFrame(x)
```

```python
x.columns = x.columns.str.strip()  # Remove leading/trailing spaces
x = x.loc[:, ~x.columns.duplicated()]  # Drop duplicate columns
```

```python
#split into training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
x_train.shape
```

```python
x.head(5)
```

```python
all(isinstance(column, str) for column in x.columns)
```

```python
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
#Over data-preprocessing from above 
```

```python
#Train a machine learning model

#for get model accuracies based on several algorithms
def getAccuracy(model):
    model.fit(x_train,y_train)
    acc=model.score(x_test,y_test)
    print(acc)
```

```python
#for linear regression

from sklearn.linear_model import LinearRegression
model=LinearRegression()
getAccuracy(model)
```

```python
#for lasso regression

from sklearn.linear_model import Lasso
lasso=Lasso()
getAccuracy(lasso)
```

```python
#for RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor() 
getAccuracy(rfr)
```

```python
#for DecisionTreeRegressor

from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
getAccuracy(model)
```

```python
#Select RandomForestRegressor and tune up to relevant hyper_parameters

from sklearn.model_selection import GridSearchCV

parameters={'n_estimators':[10,30,40,50,55,60,65,75,100],
            'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson']
           }
grid_obj=GridSearchCV(estimator=rfr,param_grid=parameters)
grid_fit=grid_obj.fit(x_train,y_train)

#get best parameters
best_model=grid_fit.best_estimator_
best_model
```

```python
#Testing with best parameters

best_model.score(x_test,y_test)
```

```python
#Save the model

import pickle
with open("predictor_model.pickle","wb") as file:
    pickle.dump(best_model,file)
```
