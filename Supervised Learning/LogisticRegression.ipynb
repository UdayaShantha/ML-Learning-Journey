```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("age_job_dataset.csv") #read .csv file to variale and from it we can get it's va
data
```

```python
plt.scatter(data.age,data.job)
```

```python
x=data[["age"]] # give values fro x axis as 2D array
x
```

```python
y=data["job"] # give y axis value as 1D array
y
```

```python
from sklearn.model_selection import train_test_split # for devide data set into trainning and testing data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8) # devide fully x and y data set into 80% for trainning and 20% for testing
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LogisticRegression # import Logistic Regression class to bulid a Logistic regression model
```

```python
model=LogisticRegression() #create a object from logistic regression class
```

```python
model.fit(x_train,y_train) #create a model from train data set or create the segment function
```

```python
y_predict_from_model=model.predict(x_test) #get predicted y values from building model by giving some x values
y_predict_from_model
```

```python
y_test
```

```python
model.score(x_test,y_test) # give the accuracy of the model with testing values (give 0-1 value)
```

```python
ages=np.array([[45],[42],[20],[30]]) # give some x values and predicat relevant y values from builting model
model.predict(ages)
```
