```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #import LinearRegression class to model/train data
```

```python
data=pd.read_csv("house_prices.csv") #read the csv file using pandas
```

```python
data
```

```python
plt.scatter(data.SquareFootage,data.Price,color="red") #show data set in a pane
plt.xlabel("Number of SquareFootage")
plt.ylabel("Total prices")
plt.title("Linear regression")
```

```python
data.SquareFootage #series of a pandas (data type)
```

```python
data.Price
```

```python
x=np.array(data.SquareFootage.values)
x
```

```python
y=np.array(data.Price.values)
y
```

```python
model=LinearRegression() #craete the object from LinearRegression class
model.fit(x.reshape(-1,1),y) # train the x and y data from the created model and the trainning data should be 2D numpy array then reshape previouysly created x 1D numpy array to 2D then y will be changed automatically
```

```python
givenX=np.array([1850]).reshape(-1,1) # craete a x value as 2D array to predicat relevan y value after trainning data set
predictY=model.predict(givenX)
print("Predict y value for given 1850 value of x is : " + str(predictY))
```

```python
#create best fit line for data set
plt.scatter(data.SquareFootage,data.Price,color="red")
plt.xlabel("Number of SquareFootage")
plt.ylabel("Total prices")
plt.title("Linear regression")
m,c=np.polyfit(x,y,1) # get m and c value for y=m*x+c
plt.plot(x,m*x+c) # draw the best fit line in data set
```

```python
c
```

```python
predictY_anotherWay=m*2300+c # get prediction value using a fit line
predictY_anotherWay
```

```python
# using multiple dependant variables of x -> create a LinearRegression Model
model.fit(data[['Bedrooms','SquareFootage','Age']],data.Price) # give 3 dependant values for x as 2D array and train data set 
```

```python
predict=model.predict([[4,1900,20]])
predict
```

```python
model.coef_ # giving the m values as y=m1*x1 + m2*x2 + m3*x3 + c for given multiple x values
```

```python
model.intercept_ # give the c value
```

```python
y_new=model.coef_[0]*4+model.coef_[1]*1900+model.coef_[2]*20+model.intercept_   # get y value from y=m1*x1 + m2*x2 + m3*x3 + c 
y_new
```
