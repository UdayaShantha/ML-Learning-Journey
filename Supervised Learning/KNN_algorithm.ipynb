```python
x=data.iloc[:,0:4]  # meke x valata gnne change krna / dena variable eka 
                    # meke cloumn index wenne 0 idn , meke iloc wlin kynne row okkma saha 0 index idn (4-1) index wenkl column wala data okkm xt assign krgnn kyla
x.head(5)
```

```python
y=data.iloc[:,-1] #meke row okkogema data gnnva anthima column eke
y.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x=scaler.fit_transform(x)
x[0:5]    #scaling x values using standardization 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier #KNN wala class ek import krnva
model=KNeighborsClassifier(n_neighbors=5)  #class eke object ekk hdnv
model.fit(x_train,y_train)   #train data tika model ekt dala train wenn denva
```

```python
predict=model.predict(x_test) #train krpui model ekt tes
predict[0:5]
```

```python
y_test[0:5]
```

```python
from sklearn.metrics import accuracy_score #impoert accuracy testing class for finding accuracy
accuracy=accuracy_score(y_test,predict)
accuracy       #get the accuracy of the model
```
