```python
import seaborn as sns
sns.pairplot(data,hue="species")    #for data visualization and get the distribution of data
```

```python
x=data.drop('species',axis=1)
x.sample(5)
```

```python
y=data['species']
y.sample(5)
```

```python
#divide data set into trainning and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#get the SVM model and train the data

from sklearn.svm import SVC
model=SVC(kernel='poly',C=5.0)
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
#considering the performance of the predicted values
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
